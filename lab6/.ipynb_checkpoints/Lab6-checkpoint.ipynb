```python
import numpy as np
import IPython
import scipy.io.wavfile
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
frate, file = scipy.io.wavfile.read("./data/wind-speech.wav")
print("Original sound")
IPython.display.display( IPython.display.Audio( file, rate=frate))
```

## Part 1

```python
# STFT code from lab 1
from numpy.fft import rfft, irfft, fftfreq
def foward_transform(input_sound, dft_size, hop_size, zero_pad, window):
    input_len = len(input_sound)
    idx = 0
    segments = []
    while idx+dft_size < input_len:
        s = input_sound[idx:idx+dft_size]
        segments.append(np.multiply(s, window))
        idx += hop_size
    else:
        s = np.append(input_sound[idx:-1], np.zeros(idx+dft_size-input_len+1))
        segments.append(np.multiply(s, window))
    frames = np.array(segments)
    
    dft_frames = []
    for idx, x in enumerate(frames):
        dft_frames.append(rfft(x, dft_size + zero_pad))
    dft_frames = np.array(dft_frames, dtype=complex)
    return dft_frames 

def inverse_transform(input_sound, dft_size, hop_size, zero_pad, window):
    leng = hop_size*(len(input_sound)-1)+dft_size+zero_pad
    output = np.zeros(leng, dtype=complex)
    for idx, row in enumerate(input_sound):
        frame = np.multiply(irfft(row, dft_size + zero_pad), window)
        output[idx*hop_size:idx*hop_size+dft_size+zero_pad] += frame
    return output

def stft( input_sound, dft_size, hop_size, zero_pad, window):
    if input_sound.ndim == 1 and np.isreal(input_sound).all():
        return foward_transform(input_sound, dft_size, hop_size, zero_pad, window)
    else:
        return inverse_transform(input_sound, dft_size, hop_size, zero_pad, window)
    
def plot_stft( input_sound, stft, dft_size, hop_size, zero_pad, frate, title=""):
    output = np.absolute(stft)**0.3
    time = np.linspace(0, output.shape[0] * hop_size / frate,output.shape[0])
    freq = np.linspace(0, frate/2, int((dft_size+zero_pad)/2 + 1))
    plt.pcolormesh(time, freq, output.T)
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (HZ)")
    plt.title(title)
    plt.show()
```

```python
Hann_window = np.hanning(512)
dft_size = 512
hop_size = int(dft_size/4)
file_stft = stft(file, dft_size, hop_size, 0, Hann_window)
plot_stft(file, file_stft, dft_size, hop_size, 0, frate, "Original Sound")
```

### First find out the mean of the noise-only part at the beginning

```python
cutoff1 = int(4.5*frate)
cutoff2 = int(6.5*frate)
noise = file[cutoff1:cutoff2]
stft_out = stft(noise, dft_size, hop_size, 0, Hann_window)
noise_mean_stft = np.mean(np.absolute(stft_out), axis=0, keepdims=True)
```

### Perform spectral subtraction & median filtering

```python
alpha = 1.5
denoised = np.zeros(file_stft.shape)
m = np.absolute(file_stft)
p = file_stft/m
denoised = m - alpha*noise_mean_stft
denoised = denoised.clip(min=0)
plot_stft(file, denoised, dft_size, hop_size, 0, frate, "Denoised sound")
out = sig.medfilt2d(np.absolute(denoised), 3)
plot_stft(file, out, dft_size, hop_size, 0, frate, "Denoised&filtered sound")
out = out * p
out = stft(out, dft_size, hop_size, 0, Hann_window)
print("Denoised sound")
IPython.display.display( IPython.display.Audio( np.real(out), rate=frate))
print("I used an alpha value around 1.2 which could effectively suppress the noise while preserving most of the human voice.")
print("Before using the median filter, there was a significant musical noise.")
print("After I apply the medfilter with kernal_size of 3 this noise is much lighter. With kernal_size of 5 the noise will disappear, but the original sound seems to be affected.")
```
