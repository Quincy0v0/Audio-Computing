```python
import IPython
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import math
from scipy.io.wavfile import read as wavread
```

```python
def convolve(input_sound, fltr):
    arr = np.zeros((2, len(input_sound)+len(fltr[0])-1))
    arr[0] = signal.convolve(input_sound,fltr[0])
    arr[1] = signal.convolve(input_sound,fltr[1])
    return arr
```

```python
def straight():
    arr = np.zeros((2,1))
    arr[0] = 1
    arr[1] = 1
    return arr   
```

```python
speed = 330
att_rate = 0.7
```

```python
def left45(srate,d = 2, head = 20):
    a = d / math.sqrt(2)
    left = math.sqrt( d**2/2 + (a - head/2)**2)
    right = math.sqrt( d**2/2 + (a + head/2)**2)
    delay = int((right-left) / speed * srate)
    att = 1 - (1-att_rate) / math.sqrt(2)
    fltr = np.zeros((2, delay + 1))
    fltr[0][0] = 1
    fltr[1][delay] = att
    return fltr
```

```python
def right80(srate, d = 2, head = 20):
    theta = np.deg2rad(80)
    a = d * np.cos(theta)
    b = d * np.sin(theta)
    left = math.sqrt(a**2 + (b + head/2)**2)
    right = math.sqrt(a**2 + (b - head/2)**2)
    delay = int(abs(right-left) / speed * srate)
    att = 1 - (1-att_rate) * np.sin(theta)
    fltr = np.zeros((2, delay + 1))
    fltr[0][delay] = att
    fltr[1][0] = 1
    return fltr
```

```python
def plot_wav(sample_rate, input_sound, title = ""):
    IPython.display.display( IPython.display.Audio( input_sound, rate=sample_rate))
    if input_sound.ndim > 1:
        length = len(input_sound[0])
    else:
        length = len(input_sound)
    time = np.linspace(0, length/sample_rate, length)
    if input_sound.ndim  == 2:
        plt.plot(time, input_sound[0])
        plt.plot(time, input_sound[1])
    else:
        plt.plot(time, input_sound)
    plt.title(title)
    plt.xlabel("second")
    plt.show()
```

```python
def plot_fltr(sample_rate, input_filter, title = ""):
    length = len(input_filter[0])
    time = np.linspace(0, length/sample_rate, length)
    plt.plot(time, input_filter[0])
    plt.plot(time, input_filter[1])
    plt.title(title)
    plt.xlabel("second")
    plt.show()
    
```

```python
file_name = "data/crumble.wav"
rate, wav = wavread(file_name)
plot_wav(rate, wav)
```

```python
fltr_strt = straight()
plot_fltr(rate,fltr_strt)
wav_strt = convolve(wav, fltr_strt)
plot_wav(rate, wav_strt)
```

```python
fltr_45l = left45(rate, 10)
plot_fltr(rate, fltr_45l)
wav_45l = convolve(wav, fltr_45l)
plot_wav(rate, wav_45l)
```

```python
fltr_80r = right80(rate)
plot_fltr(rate, fltr_80r)
wav_80r = convolve(wav, fltr_80r)
plot_wav(rate, wav_80r)
```
