```python
import numpy as np
import IPython
import math
import matplotlib.pyplot as mp
from scipy.io.wavfile import read as wavread
from scipy.io.wavfile import write as wavwrite
import scipy.signal as ss
```

```python
def stft(input_sound, dft_size, hop_size, zero_pad, window):
    if input_sound.ndim == 1:
        sample_num = math.floor(len(input_sound)/hop_size)+1
        result = []
        for i in range(sample_num):            
            dft_sample = []
            dft_sample = input_sound[i * hop_size: min(i * hop_size + dft_size, len(input_sound)-1)]
            if len(dft_sample) < dft_size:
                dft_sample = np.append(dft_sample,np.zeros(dft_size-len(dft_sample)))            
            windowed_sample = np.multiply(dft_sample, window)
            fft_result = np.fft.rfft(windowed_sample,dft_size+zero_pad)
            result.append(fft_result)    
        return np.array(result)
    elif input_sound.ndim == 2:
        sample_num = input_sound.shape[0]
        length = (sample_num-1) * hop_size + dft_size
        result = np.zeros(length)
        add_time = np.zeros(length)
        for i in range(sample_num):
            ifft_result = np.fft.irfft(input_sound[i], dft_size + zero_pad)
            unwindowed_sample = np.multiply(ifft_result[:dft_size], window)
            start = i*hop_size
            for j in range(dft_size):
                result[start+j] += unwindowed_sample[j]
        return result
    else:
        print("invalid input")   
        return
```

```python
def plot_wav(sample_rate, input_sound, wav_title = ''):
    IPython.display.display( IPython.display.Audio( input_sound, rate=sample_rate)) 
    time = np.linspace(0, len(input_sound)/sample_rate, len(input_sound))
    mp.plot(time, input_sound)    
    mp.title(wav_title)
    mp.xlabel("second")
    mp.show()
```

```python
def plot_dft(sample_rate, dft, title = '', freq_max = 0, show = True):
    dft_r = np.absolute(dft) ** 0.3
    time_v = np.linspace(0,dft_r.shape[0] * hop_size / sample_rate,dft_r.shape[0])
    
    if (freq_max != 0):
        d_freq = (sample_rate // 2) / dft.shape[1]
        freq_number = int(freq_max // d_freq + 1)
        freq_max = (freq_number - 1) * d_freq
    else:
        freq_max = sample_rate // 2
        freq_number = dft.shape[1]
        
    freq_v = np.linspace(0, freq_max, freq_number)
    mp.title(title)
    mp.xlabel("second")
    mp.ylabel("HZ")
    mp.pcolormesh(time_v,freq_v,dft_r.T[:][0:freq_number])
    if show:
        mp.show()  
```

```python
dft_size = 4096
hop_size = dft_size // 4
zero_pad = 0
window = np.hanning(dft_size + zero_pad)
```

# Part I

```python
rate, wav = wavread("data/beauty.wav")
plot_wav(rate, wav, "original sound")
dft = stft(wav, dft_size, hop_size, zero_pad, window)
plot_dft(rate, dft, "original sound stft", 3000)
```

```python
freq_cutoff = [100, 300]
peak_cutoff = [rate // freq_cutoff[1], rate // freq_cutoff[0]]
temp = np.zeros(dft.shape[1])
pitch = np.zeros(dft.shape[0])
for t in range(dft.shape[0]):
    temp = np.fft.irfft(np.multiply(dft[t], np.conj(dft[t])), dft_size + zero_pad)
    peaks = np.add(ss.find_peaks(temp[peak_cutoff[0]:peak_cutoff[1]])[0], peak_cutoff[0])
    threshold = temp[0] * 0.7
    peak = next(filter(lambda x: temp[x] > threshold, peaks), 0)
    if (peak != 0):
        pitch[t] = rate / peak
    else:
        pitch[t] = 0            
time_pitch = np.linspace(0, len(wav)/rate, dft.shape[0])  
```

```python
plot_dft(rate, dft, "orignal sound stft with tracked pitch",freq_max = 3000, show = False)
mp.plot(time_pitch, pitch, color = "red")
mp.show()
#tracked pitch is marked with red line, it is almost same as the sound pitch
```

# Part II

```python
sample = wav[303000:445000]
plot_wav(rate, sample, "bad sound")
```

```python
b, a = ss.iirfilter(10, 350*2/rate, btype='lowpass', ftype='butter', output = 'ba')
fltrd_sample = ss.filtfilt(b, a, sample)
```

```python
k = rate / 274 / 2
zero_crossing = np.where(np.diff(np.sign(fltrd_sample)))[0]
fix_sound = np.zeros(len(sample))
offset = zero_crossing[0]
fix_sound[0:offset] = sample[0:offset]
for i in range(1, len(zero_crossing) - 1):
    temp = sample[zero_crossing[i - 1]:zero_crossing[i+1]]
    start = int((i-1) * k) + offset
    window = np.hanning(len(temp))
    fix_period = np.multiply(temp, window)
    if (start + len(temp) < len(fix_sound)):
        fix_sound[start:start+len(temp)] += fix_period
```

```python
plot_wav(rate, fix_sound, "fixed \"me\" sound")
# much better than the original one, the pitch is stable now
```
